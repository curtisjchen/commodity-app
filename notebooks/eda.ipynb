```python
df1_unique = df1.groupby('UIN').agg({
    'region': "first",
    "dpmt_num": "first",
    'cat1_desc': "first",
    'cat12_desc': "first",
    'cat123_desc': "first",
    "item_desc_1": "first",
    "item_desc_2": "first"
}).reset_index()

df1_unique['cat1_desc'] = df1_unique['cat1_desc'].apply(lambda x: x.split('-', 1)[1].strip() if pd.notnull(x) and '-' in x else x)
df1_unique['cat12_desc'] = df1_unique['cat12_desc'].apply(lambda x: x.split('-', 1)[1].strip() if pd.notnull(x) and '-' in x else x)
df1_unique['cat123_desc'] = df1_unique['cat123_desc'].apply(lambda x: x.split('-', 1)[1].strip() if pd.notnull(x) and '-' in x else x)
```

```python
df2_clean = df2[['UIN', 'Selling Country/Region', 'Department', 'Category 1 Description', 'Category 2 Description', 'Category 3 Description', 'Item Desc. 1', 'Item Desc. 2', 'Commodity']].copy()
df2_clean.columns = ['uin', 'region', 'dpmt_num', 'cat1_desc', 'cat12_desc', 'cat123_desc', 'item_desc_1', 'item_desc_2', 'commodity']
df2_clean
```

```python
def combine_features(row):
    # Use empty strings for NaNs
    r = str(row['region']).strip() if pd.notnull(row['region']) else ""
    d = re.search(r'\d+', str(row['dpmt_num'])).group() if re.search(r'\d+', str(row['dpmt_num'])) else "0"
    c1 = str(row['cat1_desc']).strip() if pd.notnull(row['cat1_desc']) else ""
    c2 = str(row['cat12_desc']).strip() if pd.notnull(row['cat12_desc']) else ""
    c3 = str(row['cat123_desc']).strip() if pd.notnull(row['cat123_desc']) else ""
    desc1 = str(row['item_desc_1']).strip()
    desc2 = str(row['item_desc_2']).strip()
    
    # We only include the label if the value exists, or keep it empty
    return f"Region: {r} | Department: {d} | Category: {c1} > {c2} > {c3} | Description: {desc1} {desc2}"
```

```python
df3_clean = df3[(df3["Parent Activity EF Label"] == "Beef") | (df3["Parent Activity EF Label"] == "Pork") | (df3["Parent Activity EF Label"] == "Chicken")].copy()
df3_clean['commodity'] = df3_clean['Parent Activity EF Label'].apply(lambda x: "Beef" if re.search(r'Beef', x, re.IGNORECASE) else ("Pork" if re.search(r'Pork', x, re.IGNORECASE) else ("Chicken" if re.search(r'Chicken', x, re.IGNORECASE) else "Other")))
```

```python
df3_clean = df3_clean[['REGION', 'ITEM_NUM', 'DEPARTMENT_NUM', 'MAJOR_CATEGORY_DESC', 'GROUP_CATEGORY_DESC', 'ITEM_CATEGORY_DESC', 'ITEM_DESC_1', 'ITEM_DESC_2', 'commodity']]
df3_clean.columns = ['region', 'item_num', 'dpmt_num', 'cat1_desc', 'cat12_desc', 'cat123_desc', 'item_desc_1', 'item_desc_2', 'commodity']
```

```python
df2_clean = df2_clean.replace("Null", pd.NA)
df2_clean = df2_clean.dropna(subset=['cat1_desc', 'item_desc_1'])
```

```python
df4 = pd.read_csv("../data/raw/fy24_ksprotein.csv")
```

```python
df4_clean = df4[['UIN', 'REGION', 'DEPT.', 'ITEM NUM.',
        'CAT. 1 - DESC.', 'CAT. 12 - DESC.', 'CAT. 123 - DESC.', 
        'ITEM DESC. 1', 'ITEM DESC. 2', 'PRIMARY LABEL']].copy()
df4_clean.columns = ['uin', 'region', 'dpmt_num', 'item_num', 'cat1_desc', 'cat12_desc', 'cat123_desc', 'item_desc_1', 'item_desc_2', 'commodity']
df4_clean
```

```python
combined_df = pd.concat([df2_clean, df3_clean, df4_clean], ignore_index=True, join='inner')
```

```python
combined_df
```

```python
combined_df['dpmt_num'] = combined_df['dpmt_num'].apply(lambda x: re.search(r'\d+', str(x)).group() if re.search(r'\d+', str(x)) else "0")
combined_df['commodity'] = combined_df['commodity'].str.strip().str.upper()
```

```python
combined_df['cat123_desc'].str.contains('-').sum() / len(combined_df)
```

```python
combined_df = combined_df.dropna(subset=['cat1_desc', 'cat12_desc', 'item_desc_1']).copy()
def clean_category_split(cat_string):
    # Handle NaN or null values safely
    if not isinstance(cat_string, str):
        return ""
        
    # Split on the first hyphen, grab the last resulting item, and strip whitespace
    return cat_string.split('-', 1)[-1].strip()
for col in ['cat1_desc', 'cat12_desc', 'cat123_desc']:
    combined_df[col] = combined_df[col].apply(clean_category_split)

combined_df['concatenated_text'] = combined_df.apply(combine_features, axis=1)
combined_df['commodity'] = combined_df['commodity'].str.replace('CHICKEN', 'POULTRY')
```

```python
combined_df.to_csv("../data/processed/combined_commodity_data.csv", index=False)
```
